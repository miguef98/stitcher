# Primera version solo blanco y negro

```python
imagenA = cv2.cvtColor(cv2.imread("Imagenes/Histogram Matching/alvarez.jpg"), cv2.COLOR_BGR2GRAY)
imagenB = cv2.cvtColor(cv2.imread("Imagenes/Histogram Matching/verstappen.jpeg"), cv2.COLOR_BGR2GRAY)

plt.imshow(imagenA, cmap='gray', vmin=0, vmax=255)
```

```python
plt.imshow(imagenB, cmap='gray', vmin=0, vmax=255)
```

calculamos histogramas

```python
histSize = 256
histRange = (0, 256) # the upper boundary is exclusive
histograma_A = np.reshape(cv2.calcHist(imagenA, [0], None, [histSize], histRange), (1,256))[0]

plt.figure(0)
plt.bar(np.arange(0,256), histograma_A)
plt.show()
```

```python
histograma_B = np.reshape(cv2.calcHist(imagenB, [0], None, [histSize], histRange), (1,256))[0]

plt.figure(0)
plt.bar(np.arange(0,256), histograma_B)
plt.show()
```

calculamos las distribuciones acumuladas de los histogramas

```python
acc_histo_A = np.add.accumulate(histograma_A)
plt.figure(0)
plt.plot(np.arange(0,256), acc_histo_A, marker='None')
plt.show()
```

```python
acc_histo_B = np.add.accumulate(histograma_B)
plt.figure(0)
plt.plot(np.arange(0,256), acc_histo_B, marker='None')
plt.show()
```

Queremos que el histograma de la imagen B (verstappen) matchee con el histograma de la imagen A (alvarez)

```python
# funcion de matcheo
M = np.zeros(256)

#Then for each gray level G1 in [0,255], we find the gray level G2 for which acc_B(G1)=acc_A(G2)
for G1 in range(256):
    G2 = np.searchsorted( acc_histo_A, acc_histo_B[G1], side="left" )
    M[G1] = G2

imagenMatcheada = [ [ min(M[pixel], 255) for pixel in fila ] for fila in imagenB ] 
```

```python
plt.imshow(imagenMatcheada, cmap='gray', vmin=0, vmax=255)
```

# Version a color

```python
%reset -f
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
imagenB = cv2.imread("Imagenes/Histogram Matching/alvarez.jpg")
imagenA = cv2.imread("Imagenes/Histogram Matching/verstappen.jpeg")

plt.imshow(imagenA[...,::-1])
```

```python
plt.imshow(imagenB[...,::-1])
```

```python
histSize = 256
histRange = (0, 256) # the upper boundary is exclusive

# histogramas 
bgr_canales_A = cv2.split(imagenA)
histogramas_A = [ np.reshape(cv2.calcHist(bgr_canales_A, [i], None, [histSize], histRange), (1,256))[0] for i in range(3) ]

bgr_canales_B = cv2.split(imagenB)
histogramas_B = [ np.reshape(cv2.calcHist(bgr_canales_B, [i], None, [histSize], histRange), (1,256))[0] for i in range(3) ]

# acumuladas
acc_histo_A = [ np.add.accumulate(histograma) for histograma in histogramas_A ]
acc_histo_B = [ np.add.accumulate(histograma) for histograma in histogramas_B ]
```

```python
# funciones de matcheo
M = np.zeros( (3,256), dtype=np.uint8 )

#Then for each gray level G1 in [0,255], we find the gray level G2 for which acc_B(G1)=acc_A(G2)
for G1 in range(256):
    G2_b = np.searchsorted( acc_histo_A[0], acc_histo_B[0][G1], side="left" )
    M[0][G1] = int(min(G2_b, 255))
    
    G2_g = np.searchsorted( acc_histo_A[1], acc_histo_B[1][G1], side="left" )
    M[1][G1] = int(min(G2_g, 255))
    
    G2_r = np.searchsorted( acc_histo_A[2], acc_histo_B[2][G1], side="left" )
    M[2][G1] = int(min(G2_r, 255))

imagenMatcheada = np.array([ [ [ M[canal][int(pixel[canal])] for canal in range(3) ] for pixel in fila ] for fila in imagenB ] )
print(imagenB.shape, imagenMatcheada.shape)
```

```python
plt.imshow(imagenMatcheada[...,::-1])
```

```python
bgr_canales_C = cv2.split(imagenMatcheada)
histogramas_C = [ np.reshape(cv2.calcHist(bgr_canales_C, [i], None, [histSize], histRange), (1,256))[0] for i in range(3) ]
acc_histo_C = [ np.add.accumulate(histograma) for histograma in histogramas_C ]
```

```python
plt.figure(0)
plt.plot(np.arange(0,256), acc_histo_B[0], marker='None', color='blue')
plt.plot(np.arange(0,256), acc_histo_C[0], marker='None', linestyle='--',color='blue')
plt.plot(np.arange(0,256), acc_histo_B[1], marker='None', color='green')
plt.plot(np.arange(0,256), acc_histo_C[1], marker='None', linestyle='--',color='green')
plt.plot(np.arange(0,256), acc_histo_B[2], marker='None', color='red')
plt.plot(np.arange(0,256), acc_histo_C[2], marker='None', linestyle='--',color='red')
plt.show()
```

# Matcheando imagenes similares

```python
%reset -f
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
imagenA = cv2.imread("Imagenes/Histogram Matching/eiffel1.jpg")
imagenB = cv2.imread("Imagenes/Histogram Matching/eiffel2.jpg")

histSize = 256
histRange = (0, 256) # the upper boundary is exclusive

# histogramas 
bgr_canales_A = cv2.split(imagenA)
histogramas_A = [ np.reshape(cv2.calcHist(bgr_canales_A, [i], None, [histSize], histRange), (1,256))[0] for i in range(3) ]

bgr_canales_B = cv2.split(imagenB)
histogramas_B = [ np.reshape(cv2.calcHist(bgr_canales_B, [i], None, [histSize], histRange), (1,256))[0] for i in range(3) ]

# acumuladas
acc_histo_A = [ np.add.accumulate(histograma) for histograma in histogramas_A ]
acc_histo_B = [ np.add.accumulate(histograma) for histograma in histogramas_B ]

# funciones de matcheo
M = np.zeros( (3,256), dtype=np.uint8 )

#Then for each gray level G1 in [0,255], we find the gray level G2 for which acc_B(G1)=acc_A(G2)
for G1 in range(256):
    G2_b = np.searchsorted( acc_histo_A[0], acc_histo_B[0][G1], side="left" )
    M[0][G1] = int(min(G2_b, 255))
    
    G2_g = np.searchsorted( acc_histo_A[1], acc_histo_B[1][G1], side="left" )
    M[1][G1] = int(min(G2_g, 255))
    
    G2_r = np.searchsorted( acc_histo_A[2], acc_histo_B[2][G1], side="left" )
    M[2][G1] = int(min(G2_r, 255))

imagenMatcheada = np.array([ [ [ M[canal][int(pixel[canal])] for canal in range(3) ] for pixel in fila ] for fila in imagenB ] )

plt.imshow(imagenA[...,::-1])
```

```python
plt.imshow(imagenB[...,::-1])
```

```python
plt.imshow(imagenMatcheada[...,::-1])
```
